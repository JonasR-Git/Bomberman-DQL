```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
analysis_file = 'mean_returns_1.npy'
```

```python
avg_rewards_list = np.load(analysis_file)[-100:]
```

```python
ep = [i  for i in range(len(avg_rewards_list))]
plt.plot( range(len(avg_rewards_list)),avg_rewards_list,'b')
# include linear trend line
z = np.polyfit(ep, avg_rewards_list, 1)
p = np.poly1d(z)
plt.plot(ep,p(ep),"r--")
plt.title("Avg Test Aeward Vs Test Episods")
plt.xlabel("Test Episods")
plt.ylabel("Average Test Reward")
plt.grid(True)
plt.show()
```
